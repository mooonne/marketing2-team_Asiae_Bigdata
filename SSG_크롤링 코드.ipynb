```python
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
'http://www.ssg.com/item/itemView.ssg?itemId=1000018738050&siteNo=6001&salestrNo=2034&tlidSrchWd=%EA%B3%B1%EC%B0%BD&srchPgNo=1&src_area=ssglist'
import re
import requests
import xml.etree.ElementTree as ET

path = 'C:/Users/user/Documents/team/chromedriver/chromedriver.exe'
options = webdriver.ChromeOptions()
options.add_argument("start-maximized")
options.add_argument('disable-infobars')

#driver = webdriver.Chrome(chrome_options=options,executable_path=path)
driver = webdriver.Chrome(path)

time.sleep(1)
comm_list = []
date_list = []
star_list = []
try:
    for page in range(2810+1):
        url = 'http://www.ssg.com/item/ajaxItemCommentList.ssg?itemId=1000018738050&siteNo=6001&dualzSiteNo=7009&filterCol=10&sortCol=01&recomAttrGrpId=&recomAttrId=&page='+str(page)+'&pageSize=10'
        driver.get(url)
        time.sleep(1)
        #text=driver.page_source
        #print(text)
        root = ET.fromstring(driver.page_source)

        for comm in root.iter('postng-cntt'):
            if comm.text!=None:
                comm_list.append(comm.text)
        for date in root.iter('wrt-dt'):
            date_list.append(date.text)
        for star in root.iter('recom-eval-scr'):
            star_list.append(star.text)
        print(page, '페이지 완료!')
    for i in range(len(date_list)):
        c = comm_list[i]
        d = date_list[i]
        s = star_list[i]
        print(c, d, s)
        print('-'*50)
except NoSuchElementException:
    print('No element found')

time.sleep(15)
#driver.close()
```
